```python
import numpy as np
import PIL
from PIL import Image
```

```python
# Download an image
!gdown 1rIFD0_D0UzO6HjT5dsz2pgCXqf_GPRb2

# Read image
amos_path = "/content/amos.PNG"
amos = Image.open(amos_path)
amos = amos.resize((150,150))
amos
```

# Create contact sheet with different brightness

```python
# Import image enhancer
from PIL import ImageEnhance

# Create and fit enhancer instance
enhancer = ImageEnhance.Brightness(amos)

# Create 10 images with varying brightness
images = []
for i in range(10):
    images.append(enhancer.enhance(i/10))
```

```python
# Take a look at one of the images.
images[3]
```

```python
# Create contact sheet
sheet = Image.new(amos.mode, (amos.width*5, amos.height*2))
sheet
```

```python
# Copy the 10 images in the contact sheet
x = 0
y = 0
for image in images:
    sheet.paste(image, (x, y))

    if x + amos.width == sheet.width:
        x = 0
        y = y + amos.height
    else:
        x = x + amos.width
```

# Change Amos' color
Change the intensity of the channels in relationship to the other channels.

```python
amos
```

```python
images=[]
for i in range(3):
    for j in [0.1, 0.5, 0.9]:
        source = amos.split()
        mid = source[i].point(lambda x:x*j)
        source[i].paste(mid)
        im = Image.merge(amos.mode, source)
        images.append(im)
```

```python
# Create contact sheet
sheet = Image.new(amos.mode, (amos.width*3, amos.height*3))
```

```python
# Copy the 10 images in the contact sheet
x = 0
y = 0
for image in images:
    sheet.paste(image, (x, y))

    if x + amos.width == sheet.width:
        x = 0
        y = y + amos.height
    else:
        x = x + amos.width
```
