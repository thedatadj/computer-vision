# Make a contact sheet in PILLOW

```python
# Import PILLOW
import PIL
```

```python
# To handle images
from PIL import Image
```

```python
# Download an image
! gdown --id 1hOGlzfhJC7aBLkTOr7bfoxPJRVq5se6s
```

```python
# Read image
image_path = "/content/cat.png"
image = Image.open(image_path)
```

```python
# Display the image
image
```

```python
# For the contact sheet maker
mode = image.mode
w = image.width
h = image.height
```

```python
# Create contact sheet
sheet = PIL.Image.new(mode, (w*5, h*2))
sheet
```

```python
# Place image in contact sheet
sheet.paste(image, (0, 0))
sheet
```

```python
# Copy image 10 times
sheet = PIL.Image.new(mode, (w*5, h*2))
for row in range(2):
    for column in range(5):
        sheet.paste(image, (column*w, row*h))
sheet
```

## Create 10 variations of the image

```python
# This object will apply transformations to the image
from PIL import ImageEnhance
```

```python
# Change the brighness of the image
enhancer = ImageEnhance.Brightness(image)

# List of images
images = []


for i in range(10):
    new_image = enhancer.enhance(i/10)
    images.append(new_image)
```

```python
images
```

```python
# Copy the 10 images in a sheet
sheet = PIL.Image.new(mode, (w*5, h*2))
i = 0
for row in range(2):
    for column in range(5):
        sheet.paste(images[i], (column*w, row*h))
        i += 1
sheet
```
